```python
import re
import string

def clean_text(q):
    if not isinstance(q, str):
        return ""

    # Lowercase
    q = q.lower()

    # Replace symbols with words
    q = q.replace('%', ' percent ')
    q = q.replace('$', ' doller ')
    q = q.replace('₹', ' rupee ')
    q = q.replace('€', ' euro ')
    q = q.replace('@', ' at ')

    # Remove emojis
    emoji_pattern = re.compile(
        "["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        "]+",
        flags=re.UNICODE
    )
    q = emoji_pattern.sub(r'', q)

    # Remove punctuation
    q = q.translate(str.maketrans('', '', string.punctuation))

    # Remove extra whitespace
    q = re.sub(r'\s+', ' ', q).strip()

    return q
```

```python
df["video_title"] = df["video_title"].apply(clean_text) 
df["video_description"] = df["video_description"].apply(clean_text)
df["video_tags"] = df["video_tags"].apply(clean_text)
df["channel_title"] = df["channel_title"].apply(clean_text)
df["video_category_id"] = df["video_category_id"].apply(clean_text)
```

```python
TEXT_COLS = [
    "video_title",
    "video_description",
    "video_tags",
    "channel_title"
]

CAT_COLS = [
    "video_category_id",
]

NUM_COLS = [
    "log_video_duration_sec",
    "log_channel_subscriber_count",
    "log_channel_video_count",
    "log_channel_view_count",
]
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

# Combine text columns
def combine_text(df, cols):
    return df[cols].agg(" ".join, axis=1).tolist()

X_train_text = combine_text(X_train, TEXT_COLS)
X_test_text  = combine_text(X_test, TEXT_COLS)

# Encode text (GPU)
X_train_text_emb = embedder.encode(
    X_train_text,
    batch_size=32,
    show_progress_bar=True,
    convert_to_numpy=True
)

X_test_text_emb = embedder.encode(
    X_test_text,
    batch_size=32,
    show_progress_bar=True,
    convert_to_numpy=True
)

# 🔥 STANDARDIZATION (REQUIRED BY YOU)
text_scaler = StandardScaler()

X_train_text_emb = text_scaler.fit_transform(X_train_text_emb)
X_test_text_emb  = text_scaler.transform(X_test_text_emb)
```

```python
text_lr = LogisticRegression(
    max_iter=1000,
    n_jobs=-1
)

text_lr.fit(X_train_text_emb, y_train)

text_train_pred = text_lr.predict_proba(X_train_text_emb)[:, 1]
text_test_pred  = text_lr.predict_proba(X_test_text_emb)[:, 1]
```

```python
# Helper to encode a single text column
def encode_text_column(embedder, series, batch_size=256):
    texts = series.fillna("").astype(str).str.strip().tolist()
    return embedder.encode(
        texts,
        batch_size=batch_size,
        convert_to_numpy=True,
        normalize_embeddings=True,
        show_progress_bar=True
    )

# Example: title embeddings
X_train_title_emb = encode_text_column(embedder, X_train["video_title"])
X_test_title_emb  = encode_text_column(embedder, X_test["video_title"])

# Example: description embeddings
X_train_desc_emb = encode_text_column(embedder, X_train["video_description"])
X_test_desc_emb  = encode_text_column(embedder, X_test["video_description"])

# Example: tags embeddings
X_train_tags_emb = encode_text_column(embedder, X_train["video_tags"])
X_test_tags_emb  = encode_text_column(embedder, X_test["video_tags"])
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler

# Numeric
scaler = StandardScaler()
X_train_num = scaler.fit_transform(X_train[NUM_COLS])
X_test_num  = scaler.transform(X_test[NUM_COLS])

# Categorical
encoder = OneHotEncoder(handle_unknown="ignore", sparse_output=False)
X_train_cat = encoder.fit_transform(X_train[CAT_COLS])
X_test_cat  = encoder.transform(X_test[CAT_COLS])
```

```python
import numpy as np

X_train_final = np.hstack([X_train_title_emb, X_train_desc_emb, X_train_tags_emb,
                           X_train_num, X_train_cat])
X_test_final  = np.hstack([X_test_title_emb, X_test_desc_emb, X_test_tags_emb,
                           X_test_num, X_test_cat])
```

```python
from sklearn.ensemble import RandomForestClassifier, StackingClassifier
from sklearn.linear_model import LogisticRegression

# Base models
rf = RandomForestClassifier(
    n_estimators=150,
    min_samples_leaf=2,
    random_state=42,
    n_jobs=-1
)

lr = LogisticRegression(
    max_iter=1000,
    class_weight="balanced",
    n_jobs=-1
)
```

```python
# Base estimators
estimators = [
    ("rf", rf),
    ("lr", lr)
]

stack = StackingClassifier(
    estimators=estimators,
    final_estimator=LogisticRegression(max_iter=1000, class_weight="balanced"),
    stack_method="predict_proba",  # use probabilities for meta-model
    cv=3,  # faster than 5-fold
    n_jobs=-1
)
```

```python
stack.fit(X_train_final, y_train)
```

```python
y_pred_stack = stack.predict(X_test_final)

from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred_stack))
```

```python
example_input_trending = {
    "video_title": "Top 10 AI Tools You Must Try in 2026",
    "video_description": "Discover the latest AI tools revolutionizing work and productivity in 2026. Don't miss out on these must-try apps!",
    "video_tags": "AI, Artificial Intelligence, Tech, Productivity, Tools",
    "video_category_id": "Science & Technology",               # Technology / Science
    "log_video_duration_sec": 6.5,         # ~665 sec ≈ 11 minutes
    "log_channel_subscriber_count": 7.5,   # ~1800 subscribers (moderate reach)
    "log_channel_video_count": 5.0,        # ~148 videos
    "log_channel_view_count": 8.0          # ~3.0 million total views
}

user_df = pd.DataFrame([example_input_trending])

```

```python
# -------------------------------
# 2️⃣ Precompute embeddings for user input
# -------------------------------
def encode_text_columns_user(embedder, df, text_cols):
    embeddings_list = []
    for col in text_cols:
        emb = embedder.encode(
            df[col].fillna("").astype(str).str.strip().tolist(),
            batch_size=1,
            convert_to_numpy=True,
            normalize_embeddings=True,
            show_progress_bar=False
        )
        embeddings_list.append(emb)
    # Concatenate all text embeddings
    return np.hstack(embeddings_list)

user_text_emb = encode_text_columns_user(embedder, user_df, TEXT_COLS)
```

```python
# -------------------------------
# 3️⃣ Preprocess numeric + categorical features
# -------------------------------
# Numeric
user_num = scaler.transform(user_df[NUM_COLS])

# Categorical
user_cat = encoder.transform(user_df[CAT_COLS])
```

```python
user_features = np.hstack([user_text_emb, user_num, user_cat])
```

```python
user_pred = stack.predict(user_features)
user_proba = stack.predict_proba(user_features)

print("Prediction (0 = Will not trend, 1 = Will trend):", user_pred[0])
print("Prediction probabilities:", user_proba[0])
```

```python
# Access fitted Random Forest base model
rf_fitted = stack.named_estimators_["rf"]

# Access fitted Logistic Regression text model
lr_fitted = stack.named_estimators_["lr"]

# Predict on your user features
rf_pred = rf_fitted.predict(user_features)
lr_pred = lr_fitted.predict(user_features)

print("RF base model prediction:", rf_pred)
print("LR base model prediction:", lr_pred)

```

```python
SHOCKING_WORDS = {
    "shocking", "unbelievable", "insane", "exposed", "leaked",
    "truth", "scam", "fraud", "dangerous"
}

URGENCY_WORDS = {
    "now", "today", "urgent", "breaking", "warning", "alert",
    "immediately", "stop", "important"
}

EMOTION_WORDS = {
    "emotional", "heartbreaking", "sad", "angry", "happy",
    "proud", "crying", "tears", "fear", "panic"
}

HYPE_WORDS = {
    "viral", "massive", "huge", "biggest", "crazy",
    "record", "explodes", "trending"
}

YOUTUBE_HOOK_WORDS = {
    "you wont believe", "watch till end", "this happened",
    "what happened next", "must watch", "dont miss"
}
```

```python
df["combined_text"] = (
    df["video_title"].fillna("") + " " +
    df["video_description"].fillna("") + " " +
    df["video_tags"].fillna("")
).str.lower()
```

```python
def has_any_word(text, word_set):
    if not text:
        return 0
    return int(any(w in text for w in word_set))
```

```python
df["has_shocking"] = df["combined_text"].apply(
    lambda x: has_any_word(x, SHOCKING_WORDS)
)

df["has_urgency"] = df["combined_text"].apply(
    lambda x: has_any_word(x, URGENCY_WORDS)
)

df["has_emotion"] = df["combined_text"].apply(
    lambda x: has_any_word(x, EMOTION_WORDS)
)

df["has_hype"] = df["combined_text"].apply(
    lambda x: has_any_word(x, HYPE_WORDS)
)

df["has_hook"] = df["combined_text"].apply(
    lambda x: has_any_word(x, YOUTUBE_HOOK_WORDS)
)
```

```python
total = len(df)

ratios = {
    "shocking_ratio": df["has_shocking"].sum() / total,
    "urgency_ratio": df["has_urgency"].sum() / total,
    "emotion_ratio": df["has_emotion"].sum() / total,
    "hype_ratio": df["has_hype"].sum() / total,
    "hook_ratio": df["has_hook"].sum() / total
}

ratios
```
